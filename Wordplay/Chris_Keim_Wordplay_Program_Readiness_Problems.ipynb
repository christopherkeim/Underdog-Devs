# 1. What are all of the words containing UU?

# 10. What are all of the words that have all 5 vowels, in alphabetical order?

```python
"""
Part 1: extract the words that have all 5 vowels
"""
# Initialize all_vowels list
all_vowels = []

# Loop through each string in scrabble_words
for word in scrabble_words:
    # Check if the string contains any vowels
    if 'A' in word and 'E' in word and 'I' in word and 'O' in word and 'U' in word:
        all_vowels.append(word)

"""
Part 2: Extract only the words that have all 5 vowels in alphabetical order from this list
"""
# Define a list of vowels
vowels = ['A', 'E', 'I', 'O', 'U']

# Initialize our sorted_vowels list
sorted_vowels = []

# Loop through each string that contains all 5 vowels
for word in all_vowels:
    # Initialize intra_word_vowels list
    intra_word_vowels = []

    # Loop through each char in this string
    for c in word:
        # Check if this char is a vowel
        if c in vowels:
            # Save it into intra_word_vowels
            intra_word_vowels.append(c)

    # Check if the intra_word_vowels list is in alphabetical order, and if so add this word to sorted_vowels
    if intra_word_vowels == sorted(intra_word_vowels):
        sorted_vowels.append(word)

# Loop for readability
for w in sorted_vowels:
    print(w)
```

**Note:** This was a bit of a doozy. In this implementation, I tried to capture edge cases as I was thinking about this into the scope of my code, as in:

> What happens if there are two consecutive vowels that are the same, as in "a" followed by another "a", or two consecutive "e" vowels within the word?

# 3. What is the shortest word that contains all 5 vowels?

```python
# Initialize all_vowels list
all_vowels = []

# Loop through each string in scrabble_words
for word in scrabble_words:
    # Check if the string contains all 5 vowels
    if 'A' in word and 'E' in word and 'I' in word and 'O' in word and 'U' in word:
        all_vowels.append(word)

# Initialize shortest_length with a high value and shortest_word as None
shortest_length = 100
shortest_word = None

# Loop through each word that contains all 5 vowels
for word in all_vowels:
    # If the length of this word is smaller than the shortest_length
    if len(word) < shortest_length:
        # Update shortest_length
        shortest_length = len(word)
        # Update the shortest_word
        shortest_word = word

# Print the shortest word
print(shortest_word)
```

# 4. What is the longest word that contains no vowels?

```python
# Initialize all_vowels list
no_vowels = []

# Loop through each string in scrabble_words
for word in scrabble_words:
    # Check if the string contains no vowels
    if 'A' not in word and 'E' not in word and 'I' not in word and 'O' not in word and 'U' not in word:
        no_vowels.append(word)

# Initialize longest_length with a small value and longest_word as None
longest_length = 0
longest_word = None

# Loop through each word that contains no vowels
for word in no_vowels:
    # If the length of this word is longer than the longest_length
    if len(word) > longest_length:
        # Update longest_length
        longest_length = len(word)
        # Update the longest_word
        longest_word = word

# Print the shortest word
print(longest_word)
```

# 5. Which of the letters Q, X, and Z is the least common?

```python
# Initialize our qxz dictionary
qxz = {'Q': 0, 'X': 0, 'Z': 0}

# Loop through each string in scrabble_words
for word in scrabble_words:
    # Check if the string contains each letter and incremement their count
    for char in word:
        if char in qxz:
            qxz[char] += 1


# Keep this scalable if we were measuring the frequencies of more letters
least_common = 1000000
letter = None

# Loop through each key
for key in qxz:
    # If this key's count is smaller than the least_common
    if qxz[key] < least_common:
        # Update least_common
        least_common = qxz[key]
        # Update our least common letter
        letter = key

# Print this letter
print(letter)
```

# 6. What is the longest palindrome?

**Defintion of a palindrome:** a word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run.

```python
# Initialize pals list
pals = []

# Loop through each string in scrabble_words
for word in scrabble_words:
    # Check if the string is a palindrome
    if word == word[::-1]:
        pals.append(word)

# Initialize longest_length with a small value and longest_pal as None
longest_length = 0
longest_pal = None

# Loop through each palindrome
for p in pals:
    # If the length of this palindrome is greater than the longest_length
    if len(p) > longest_length:
        # Update longest_length
        longest_length = len(p)
        # Update the longest_pal
        longest_pal = p

# Print the longest palindrome
print(longest_pal)
```

# 7. What are all of the letters that never appear consecutively in an English word? For example, we know that U isnt an answer, because of the word VACUUM, and we know that A isnt an answer, because of AARDVARK, but which letters never appear consecutively?

```python
# Define our letters
letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

# Initialize nev_consec dictionary
consec_letters = {}

# Populate our dictionary with letter keys each holding the value False
for l in letters:
    consec_letters[l] = False

# Loop through each string in scrabble_words
for word in scrabble_words:
    # Loop through each letter in this word
    for i in range(len(word) - 1):
        # Check to see if this letter appears consecutively with the one following letter
        if word[i] == word[i+1]:
            consec_letters[word[i]] = True

# Extract the letter keys with the value False
nev_consec = []
for key in consec_letters:
    if consec_letters[key] == False:
        nev_consec.append(key)

# Print the letters that never occur consecutively
print(nev_consec)
```

An interesting problem - my approach here was to assume the hypothesis that each letter does not appear consecutively in any English words. Then, I ran a logical test to disprove this for each character in the file. Any character suriving this test (meaning its original value of False was unchanged) can be assumed to "not occur consecutively in a English word." - Though at some intuitive level I don't like this kind of negative evidence.
